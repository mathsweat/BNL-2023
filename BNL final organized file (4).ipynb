```python
def euclidean_distance(x,y,a,b):
    d=((a-x)**2+(b-y)**2)**0.5
    return d


def cluster_avg(val_list):
    hold_avgs=np.array([])
    hold_indices=np.array([])
    lastindex=0
    for j in range(len(val_list)-1):#we don't want to go out of bounds. so keep the j+1 from going too far.
        if(val_list[j+1]>(val_list[j]+10)): #The +10 is just a number I have to be the amount above to split. So if values differ by 10, the array splits between these indices.
            for k in range(lastindex, (j+1)): #go from previous cluster's last index to the next one that starts a new cluster.
                hold_indices=np.append(hold_indices, val_list[k]) #put the cluster's indices in an array
            hold_avgs=np.append(hold_avgs, np.nanmean(hold_indices)) #put the mean of the cluster's indices in an array
            hold_indices=np.array([]) #empty the array that holds the indices
            lastindex=j+1 # move the last index up so we will not look at the same cluster again in our for loop
    
        elif(val_list[j+1]==val_list[-1]):
            for k in range(lastindex, (j+1)): #go from previous cluster's last index to the next one that starts a new cluster. This doesn't include the final array element
                hold_indices=np.append(hold_indices, val_list[k]) #put the cluster's indices in an array
            hold_indices=np.append(hold_indices, val_list[-1]) #include the final element of save
            hold_avgs=np.append(hold_avgs, np.nanmean(hold_indices)) #put the mean of the cluster's indices in an array
            hold_indices=np.array([]) #empty the array that holds the indices
            lastindex=j+1 # move the last index up so we will not look at the same cluster again in our for loop
    avg_of_clusters=np.nanmean(hold_avgs)  #the average of the clusters we got in the row or column above the threshold
    
    return(avg_of_clusters)


def two_dim_cluster_avg(val_list_x,val_list_y): #we only compare elements next to each other becuase for threshold values far from each other we shouldn't be getting similar centers
    hold_avgs_x=np.array([])
    hold_avgs_y=np.array([])
    hold_indices_x=np.array([])
    hold_indices_y=np.array([])
    lastindex=0
    for j in range(len(val_list_x)-1): #cluster_avg takes whole list in a rows or column and makes clusters then gets the average of them. This function outputs coordinates of the average of clustered groups.
        if(euclidean_distance(val_list_x[j],val_list_y[j],val_list_x[j+1],val_list_y[j+1])>30): #no reason for 30, I just think this is a good amount
            for k in range(lastindex, (j+1)): #go from previous cluster's last index to the next one that starts a new cluster.
                hold_indices_x=np.append(hold_indices_x, val_list_x[k]) #put the cluster's indices in an array
                hold_indices_y=np.append(hold_indices_y, val_list_y[k])
            hold_avgs_x=np.append(hold_avgs_x, np.nanmean(hold_indices_x)) #put the mean of the cluster's indices in an array
            hold_avgs_y=np.append(hold_avgs_y, np.nanmean(hold_indices_y))
            hold_indices_x=np.array([]) #empty the array that holds the indices
            hold_indices_y=np.array([])
            lastindex=j+1 # move the last index up so we will not look at the same cluster again in our for loop
        if(euclidean_distance(val_list_x[j],val_list_y[j],val_list_x[j+1],val_list_y[j+1])==euclidean_distance(val_list_x[-2],val_list_y[-2],val_list_x[-1],val_list_y[-1])):
            for k in range(lastindex, (j+1)):
                hold_indices_x=np.append(hold_indices_x, val_list_x[k]) #put the cluster's indices in an array
                hold_indices_y=np.append(hold_indices_y, val_list_y[k])
            hold_indices_x=np.append(hold_indices_x, val_list_x[-1])
            hold_indices_y=np.append(hold_indices_y, val_list_y[-1])
            hold_avgs_x=np.append(hold_avgs_x, np.nanmean(hold_indices_x))
            hold_avgs_y=np.append(hold_avgs_y, np.nanmean(hold_indices_y))
            hold_indices_x=np.array([])
            hold_indices_y=np.array([])
            lastindex=j+1
    #we need to return both hold_indices arrays. We can't have two return statements. We will just split the array in half later
    return(np.append(hold_avgs_x,hold_avgs_y))
    


def mode_finder(list_input):
    vals,counts = np.unique(list_input, return_counts=True)
    mode = np.argmax(counts)
    true_mode=vals[mode]
    
    return(true_mode)




def one_dim_search_rows(array_input,threshold_value):
    hold_avg=np.array([])
    holding_array=np.array([])
    for a in range(len(array_input)):
        valmax=np.amax(array_input[a]) #get the actual max value in the array
        lowerbound=threshold_value*valmax #we want to see where values above this are. The high intensity values. I chose a 15% threshold.
        for i in range(len(array_input[a])): #go through one row at a time
            if img2[a][i]>lowerbound:
                holding_array=np.append(holding_array,i) #get indices of higher values in an array
        
        hold_avg=np.append(hold_avg,cluster_avg(holding_array))
        holding_array=np.array([])
        
    return(hold_avg)



    
def one_dim_search_cols(array_input,threshold_value):
    hold_avg=np.array([])
    holding_array=np.array([])
    for b in range(len(array_input[0])): #go through every column and note that img2[0] is the size of the number of columns.
        valmax=np.amax(array_input[:,b]) #get the actual max value in that column. 
        lowerbound=threshold_value*valmax #we want to see where values above this are. The high intensity values. I chose a 15% threshold.
        for j in range(len(img2[:,b])): #go through each index of the column
            if img2[:,b][j]>lowerbound: #if the element is larger than our bound
                holding_array=np.append(holding_array,j) #get the indicies of the larger elements and add them all to a list
                
        hold_avg=np.append(hold_avg,cluster_avg(holding_array))
        holding_array=np.array([])
                
    return(hold_avg)


def closest_point_index(split_array,compare_point):
    hold_index=0
    lowest_dist=euclidean_distance(split_array[0][0],split_array[1][0],compare_point[0],compare_point[1])
    for i in range(len(split_array[0])):
        c=euclidean_distance(split_array[0][i],split_array[1][i],compare_point[0],compare_point[1])
        if c<lowest_dist:
            hold_index=i
    return(hold_index)
            

#go through each row and if intensities are less than 100 up to a certain part of the row or 
#column, make all of those points equal to the row or column median. The low values are skewing our results for file 3
#so if we encounter a value above or equal to 100, make everything before it equal to the row median
def modify_array(input_array):
    x=np.nanmean(input_array)
    for i in range(len(img2)):
        if(np.all(input_array[i]==-1)):
            for b in range(len(input_array[0])):
                input_array[i][b]=x
                
                    
                
                
    return(input_array)
    
```

```python
#vary the threshold by 0.05 and plot the centers
plt.figure(100, figsize=(12, 12))
plt.imshow(np.log10(img2+3), cmap="gist_stern")
cbar=plt.colorbar(label="scalarized point brightness")

threshold=0.999
count=20

img3=img2 #we don't want to modify img2 as it will make both arrays the same
img4=modify_array(img3)

hold_keep_mode_x=np.array([])
hold_keep_mode_y=np.array([])

hold_keep_mode_x2=np.array([])
hold_keep_mode_y2=np.array([])


while(threshold>0):
    
    
    
    keep_mode=np.array([])
    keep_mode2=np.array([])


                
#for rows

    keep_mode=np.append(keep_mode, mode_finder(one_dim_search_rows(img3,threshold)))
    keep_mode2=np.append(keep_mode2, mode_finder(one_dim_search_rows(img4,threshold)))



#for columns
    
    keep_mode=np.append(keep_mode, mode_finder(one_dim_search_cols(img3,threshold)))
    keep_mode2=np.append(keep_mode2, mode_finder(one_dim_search_cols(img4,threshold)))


    
    hold_keep_mode_x=np.append(hold_keep_mode_x,keep_mode[0])
    hold_keep_mode_y=np.append(hold_keep_mode_y,keep_mode[1])
    
    
    hold_keep_mode_x2=np.append(hold_keep_mode_x2,keep_mode2[0])
    hold_keep_mode_y2=np.append(hold_keep_mode_y2,keep_mode2[1])
    
    threshold-=0.05
    
    
    
    print(count)
    count-=1 #just counting down the values to show how long the runtime is


cluster_pieces=np.array_split(two_dim_cluster_avg(hold_keep_mode_x,hold_keep_mode_y),2) #split the array into the x and y parts
cluster_pieces2=np.array_split(two_dim_cluster_avg(hold_keep_mode_x2,hold_keep_mode_y2),2)

#compare which center is closer

if(euclidean_distance(cluster_pieces[0][closest_point_index(cluster_pieces,guess)],cluster_pieces[1][closest_point_index(cluster_pieces,guess)],guess[0],guess[1])<=euclidean_distance(cluster_pieces2[0][closest_point_index(cluster_pieces2,guess)],cluster_pieces2[1][closest_point_index(cluster_pieces2,guess)],guess[0],guess[1])):
    print("Center: ["+str(cluster_pieces[0][closest_point_index(cluster_pieces,guess)])+","+str(cluster_pieces[1][closest_point_index(cluster_pieces,guess)])+"]")
    plt.scatter(cluster_pieces[0][closest_point_index(cluster_pieces,guess)], cluster_pieces[1][closest_point_index(cluster_pieces,guess)], s=15, c="blue")
else:
    print("Center: ["+str(cluster_pieces2[0][closest_point_index(cluster_pieces2,guess)])+","+str(cluster_pieces2[1][closest_point_index(cluster_pieces2,guess)])+"]")
    plt.scatter(cluster_pieces2[0][closest_point_index(cluster_pieces2,guess)], cluster_pieces2[1][closest_point_index(cluster_pieces2,guess)], s=15, c="blue")






plt.plot(guess[0],guess[1],"gx")
```
